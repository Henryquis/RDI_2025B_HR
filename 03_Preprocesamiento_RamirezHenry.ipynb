##### 1. Limpiar los documentos de caracteres que no corresponden

```python
import re

def clean_text(doc):
    # primero reemplaza los puntos por espacios
    doc = doc.replace('.', ' ')
    # luego elimina etiquetas HTML y caracteres no permitidos,
    # conservando solo los guiones que unen letras o números
    return re.sub(r'<.*?>|(?<![A-Za-z0-9])-|-(?![A-Za-z0-9])|[^A-Za-z0-9\- ]', '', doc)

```

```python
df["review"].apply(clean_text)
```
